# LMM preprocessing

## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Import data => data frame



```python
df = pd.read_csv("Smile2023_10_25_combined.csv")
df.head()
```

```python
df.describe()
```

### Empathic Concern score distribution

```python
plt.violinplot(df["EC"])
```

## Extract only three columns for practice


*   ID
*   EC
*   smile0~3, P7, P8






```python
df_N170 = df[["ID", "EC", "sex", "sex_dummy", "smile0_P7", "smile1_P7", "smile2_P7", "smile3_P7", "smile0_P8", "smile1_P8", "smile2_P8", "smile3_P8"]]
df_N170.head()
```

## Standardization =>

```python
df_N170["EC_standardized"] = (df_N170["EC"] - df_N170["EC"].mean())/df_N170["EC"].std()
```

```python
print(df_N170["EC_standardized"].mean())
print(df_N170["EC_standardized"].std())
```

## Convert to Long format

```python
df_N170_long = pd.melt(df_N170, id_vars=["ID", "EC", "EC_standardized", "sex", "sex_dummy"])
df_N170_long.head()
```

```python
conditions_P7 = [
    (df_N170_long["variable"] == "smile0_P7"),
    (df_N170_long["variable"] == "smile1_P7"),
    (df_N170_long["variable"] == "smile2_P7"),
    (df_N170_long["variable"] == "smile3_P7")
]

conditions_P8 = [
    (df_N170_long["variable"] == "smile0_P8"),
    (df_N170_long["variable"] == "smile1_P8"),
    (df_N170_long["variable"] == "smile2_P8"),
    (df_N170_long["variable"] == "smile3_P8")
]

choices_smile_num = [0, 1, 2, 3]
choices_hemisphere_P7 = ["left", "left", "left", "left"]
choices_hemisphere_P8 = ["right", "right", "right", "right"]

df_N170_long["smile_num"] = np.select(conditions_P7, choices_smile_num, default=None)
df_N170_long["hemisphere"] = np.select(conditions_P7, choices_hemisphere_P7, default=None)

df_N170_long["smile_num"] = np.select(conditions_P8, choices_smile_num, default=df_N170_long["smile_num"])
df_N170_long["hemisphere"] = np.select(conditions_P8, choices_hemisphere_P8,default=df_N170_long["hemisphere"])

df_N170_long.head()
```

```python
df_N170_long.to_csv("N170_long.csv")
```
