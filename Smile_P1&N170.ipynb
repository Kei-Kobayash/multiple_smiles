```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Smile#_120-140&180-200.csv")

df.head()
```

```python
#列名の取得

conditions = df.columns.values

#number of participants
pars_num = len(df["ERPset"].unique())

#creating lists of conditions
P7_conditions = conditions[7:15:2]
P8_conditions = conditions[8:15:2]

PO7_conditions = conditions[23:31:2]
PO8_conditions = conditions[24:31:2]

#条件リスト名を引数として渡すと、平均と標準誤差をそれぞれリストとして返す関数

def ave_sem(data, cond_list, pars_num):
  mean_list = []
  std_list = []
  for cond in cond_list:
    mean_list.append(round(data[cond].mean(), 3))
    std_list.append(round(data[cond].std(), 3))
  sem_list = list(map(lambda x:round(x/np.sqrt(pars_num), 3),std_list))

  return mean_list, sem_list

  #print(mean_list)
  #print(std_list)
  #print(sem_list)

#関数を使う

P7_ave, P7_sem = ave_sem(df, P7_conditions, pars_num)
P8_ave, P8_sem = ave_sem(df, P8_conditions, pars_num)
PO7_ave, PO7_sem = ave_sem(df, PO7_conditions, pars_num)
PO8_ave, PO8_sem = ave_sem(df, PO8_conditions, pars_num)

print(P8_conditions)
print(P7_ave)
print(PO7_conditions)
print(PO8_conditions)
```

```python
df_EC_low = df.query('EC_group == "low"')
pars_num_low = len(df_EC_low["ERPset"].unique())

P7_EC_low_ave, P7_EC_low_sem = ave_sem(df_EC_low, P7_conditions, pars_num_low)
P8_EC_low_ave, P8_EC_low_sem = ave_sem(df_EC_low, P8_conditions, pars_num_low)
PO7_EC_low_ave, PO7_EC_low_sem = ave_sem(df_EC_low, PO7_conditions, pars_num_low)
PO8_EC_low_ave, PO8_EC_low_sem = ave_sem(df_EC_low, PO8_conditions, pars_num_low)

df_EC_low.head()

#PO7_EC_low_ave
#PO8_EC_low_ave
```

```python
df_EC_high = df.query('EC_group == "high"')

pars_num_high = len(df_EC_high["ERPset"].unique())


P7_EC_high_ave, P7_EC_high_sem = ave_sem(df_EC_high, P7_conditions, pars_num_high)
P8_EC_high_ave, P8_EC_high_sem = ave_sem(df_EC_high, P8_conditions, pars_num_high)
PO7_EC_high_ave, PO7_EC_high_sem = ave_sem(df_EC_high, PO7_conditions, pars_num_high)
PO8_EC_high_ave, PO8_EC_high_sem = ave_sem(df_EC_high, PO8_conditions, pars_num_high)

df_EC_high.head()
```

```python
#2つのリストを引数として渡すとdfにhemi_subを追加する関数

hemisub_conditions = ["hemisub 0", "hemisub 1", "hemisub 2", "hemisub 3"]
hemisub_P1_conditions = ["hemisub P1 0", "hemisub P1 1", "hemisub P1 2", "hemisub P1 3"]

def hemisub(data, left_hem_list, right_hem_list, results_conds):
  for results_cond, right_hem, left_hem in zip(results_conds, right_hem_list, left_hem_list):
    data[results_cond] = data[right_hem] - data[left_hem]


hemisub(df, P7_conditions, P8_conditions, hemisub_conditions)
hemisub(df, PO7_conditions, PO8_conditions, hemisub_P1_conditions)

hemisub_ave, hemisub_sem = ave_sem(df, hemisub_conditions, pars_num)


hemisub_P1_EC_high_ave, hemisub_P1_EC_high_sem = ave_sem(df.query('EC_group == "high"'), hemisub_P1_conditions, pars_num)
hemisub_P1_EC_low_ave, hemisub_P1_EC_low_sem = ave_sem(df.query('EC_group == "low"'), hemisub_P1_conditions, pars_num)

print(hemisub_P1_EC_high_ave)
```

```python
##散布図を描く

def scatter(var1, var2, data):
  fontsize = 12
  fig, ax1 = plt.subplots()
  ax1.scatter(data[var1], data[var2], c="black", alpha=1, edgecolors="gray", s = 50)

  ax1.spines["top"].set_visible(False)
  ax1.spines["right"].set_visible(False)

  ax1.set_xlabel("EC score", fontsize = fontsize)
  ax1.set_ylabel("Amplitude difference (μV)", fontsize = fontsize)
  ax1.set_title("3 smiles - 0 smile (N170) at P7", fontsize = fontsize)

  ax1.set_xticks(list(range(15, 36, 5)))
  ax1.set_xticklabels(list(range(15, 36, 5)), fontsize = fontsize)
  ax1.set_yticks(list(range(-2, 5, 1)))
  ax1.set_yticklabels(list(range(-2, 5, 1)), fontsize = fontsize)

  import numpy as np
  # 回帰直線の計算
  coefficients = np.polyfit(data[var1], data[var2], deg=1)
  x_values = np.linspace(data[var1].min(), data[var1].max(), 100)
  y_values = np.polyval(coefficients, x_values)

  # トレンドラインの描画
  ax1.plot(x_values, y_values, color='gray', label='トレンドライン')


scatter("EC", "smile3-0_P7", df)
```
